<h1 style="font-size:300%;"> dalla tabella dba__reparti creo un dataframe formato dall'id di ogni reparto e dal reparto

```python
import MySQLdb
import datetime
import pandas as pd

# Open database connection
db = MySQLdb.connect("localhost","root","paperino","database_stragi" )

# prepare a cursor object using cursor() method
cursor = db.cursor()

sql = '''
SELECT REP.id_reparto, REP.reparto
FROM dba__reparti AS REP
'''
#il SELECT con meno di 2 colonne da' errore

cursor.execute(sql)  
results = cursor.fetchall()

lista=list(results) 

df1 = pd.DataFrame(lista,columns=('id_reparto','nazisti/fascisti'))                    
```

```python
df1.head()
```

<h1 style="font-size:300%;">Creo la series "id_reparti_series" formata dall'ID che identifica ogni reparto

```python
id_reparti_series=df1['id_reparto']
id_reparti_series.head()
```

<h1 style="font-size:300%;"> Creo il dataframe su cui applicherò l'algoritmo Apriori

## Creo un dataframe le cui  colonne sono rappresentate dalla series "id_reparti_series" e con un numero di righe pari a quella del dataframe df_reparti

```python
import pandas as pd
import numpy as np
df_per_regole_associative = pd.DataFrame(0, index=np.arange(df_reparti.shape[0]), columns=id_reparti_series)
df_per_regole_associative.tail()
```

Il dataframe sopra, ha un numero di righe pari al numero delle stragi ed un numero di colonne pari ai vari reparti coinvolti nelle stragi.
E' però un dataframe formato da soli 0.

Per aggiungere dei valori diversi da 0, facciamo una scansione del dataframe reparti coinvolti nelle stragi e, ogni volta che incontriamo l'ID di un reparto aggiungiamo un 1 alla colonna del dataframe df_per_regole_associative nella riga relativa alla strage considerata

```python
import re
# lavoro su ogni riga del dataframe reparti
for indice_riga,riga in df_reparti.iterrows():
    # lavoro su ogni valore di ogni riga del 
    # dataframe reparti
    for id_reparto in riga:
        #df_per_regole_associative.loc[0][1291]=1
        # uso le regex per verificare quando la riga 
        # è un numero
        #print('prima del regex: ',id_reparto)
        if id_reparto is None:
            break
        stringa_match=re.match(r'[0-9]+',id_reparto)
        if stringa_match:
            #print(indice_riga,id_reparto)
            #df_per_regole_associative.loc[0][1291]=1
            #df_per_regole_associative[id_reparto].iloc[indice_riga]=1
            #df_per_regole_associative=df_per_regole_associative.loc[indice_riga][id_reparto]=1
            df_per_regole_associative.loc[int(indice_riga)][int(id_reparto)]=1
            # df.at['C', 'x'] = 10
            #df_per_regole_associative.at['indice_riga','id_reparto']=1
            #df.xs('C')['x']=10
            #df_per_regole_associative.xs('indice_riga')['id_reparto']=1
df_per_regole_associative.head(15)
```

<h1 style="font-size:300%;"> Algoritmo APRIORI: devo assegnare il supporto minimo 

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(df_per_regole_associative, min_support=0.0003,use_colnames=True)
frequent_itemsets.head()
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
rules.head()
```

```python
import MySQLdb
import datetime
import pandas as pd

# Open database connection
db = MySQLdb.connect("localhost","root","paperino","database_stragi" )

# prepare a cursor object using cursor() method
cursor = db.cursor()

sql = '''
SELECT REP1.* 
FROM dba__reparti AS REP1
WHERE REP1.id_reparto IN (204,822,823,6,223,15,579)
'''


#sql = 'SELECT REP1.*' + 'FROM dba__reparti AS REP1' + 'WHERE REP1.id_reparto IN ' + '(204,223,579,328)'


cursor.execute(sql)  
results = cursor.fetchall()

lista=list(results) 
df_reparti_associati = pd.DataFrame(lista,columns=[i[0] for i in cursor.description])
df_reparti_associati
```
