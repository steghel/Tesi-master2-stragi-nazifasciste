```python
import MySQLdb
import datetime
import pandas as pd

# Open database connection
db = MySQLdb.connect("localhost","root","paperino","database_stragi" )

# prepare a cursor object using cursor() method
cursor = db.cursor()

sql = '''
SELECT ST.datainizio, CO.lat, CO.lng
FROM dba__stragi AS ST JOIN dba__coordinate AS CO ON ST.id_strage=CO.id_strage
ORDER BY ST.datainizio 
'''

cursor.execute(sql)  
results = cursor.fetchall()

lista=list(results) 

df = pd.DataFrame(lista,columns=('data_inizio','latitudine','longitudine'))
```

```python
df.head()
```

```python
for indice,riga in df.iterrows():
    if riga.data_inizio==None:
        df=df.drop([indice]) # tolgo la riga e aggiorno il dataframe
    if riga.data_inizio==datetime.date(1, 1, 1):
        df=df.drop([indice]) # tolgo la riga e aggiorno il dataframe
    if riga.latitudine=="43.353583,":
        df=df.drop([indice])
        
```

<H1 style="cfont-family: verdana;  font-size: 60px;">Intervalli temporali in cui si verifica il numero massimo di eventi

```python
import MySQLdb
import datetime
import pandas as pd

# Open database connection
db = MySQLdb.connect("localhost","root","paperino","database_stragi" )

# prepare a cursor object using cursor() method
cursor = db.cursor()

sql = '''
SELECT datainizio,COUNT(datainizio) AS eventi
FROM dba__stragi
GROUP BY datainizio
ORDER BY datainizio
'''
cursor.execute(sql)  
results = cursor.fetchall()

lista=list(results) 
```

```python
df1 = pd.DataFrame(lista,columns=('data_inizio','numero_eventi'))
```

```python
# pulisco i dati
for indice,riga in df1.iterrows():
    if riga.data_inizio==None:
        df1=df1.drop([indice]) # tolgo la riga e aggiorno il dataframe
    if riga.data_inizio==datetime.date(1, 1, 1):
        df1=df1.drop([indice]) # tolgo la riga e aggiorno il dataframe
df1.head()
```

```python
df1.numero_eventi.sum()
```

```python
df1.shape[0]
```

```python
# faccio iniziare l'indice delle righe da 0
df1.index=range(df1.shape[0])
df1.head()
```

```python
df1.describe()
```

```python
df1.shape[0]
```

```python
# lista degli intervalli temporali in cui abbiamo il numero massimo di eventi
import numpy as np
import pandas as pd

lista=[]
coppia_inizio_fine=[]
parola='inizio'
parziale=0

for indice,riga in df1.iterrows():
    #print(riga['numero_eventi'])
    if abs(riga['numero_eventi'])>20 and parola=='inizio':
        #print(indice,(df1.iloc[indice])['data_inizio'])
        parziale=(df1.iloc[indice])['data_inizio']
        lp=[]
        fine=indice+10
        if fine>df1.shape[0]:
            fine=df1.shape[0]
        for i in range(indice,fine):
            
            #print('valore',df1.iloc[i])
            lp.append((df1.iloc[i])['numero_eventi'])
        media=np.mean(lp) 
        #print('media',media)
        if media>10:
            
            coppia_inizio_fine.append(parziale)
            #print(coppia_inizio_fine)
            #parziale=0
            parola='fine'
        #print(parola)
        
    if abs(riga['numero_eventi'])<10 and parola=='fine':
        #print('ok')
        parziale=(df1.iloc[indice])['data_inizio']
        lp1=[]
        fine=indice+10
        if fine>df1.shape[0]:
            fine=df1.shape[0]
        for i in range(indice,fine):
            #print(indice,indice+100)
            #lp1.append(abs(riga['canale_sin'])) 
            lp1.append((df1.iloc[i])['numero_eventi'])
        media1=np.mean(lp1) 
        #print(media1)
        if media1<8:
            coppia_inizio_fine.append(parziale)
            lista.append(coppia_inizio_fine)
            coppia_inizio_fine=[]
            parziale=0
            parola='inizio'  
lista
```

#### PROBLEMA 
Ho problemi a trovare i cluster tra data, latitud e longit perchè abbiamo a che fare con dati aventi scale diverse:<div></div>il valore dei giorni varia tra 0 e 683 mentre lat e longit variano in una finestra ("valore_massimo" , "valore_minimo") di  valori interi.<div></div>
#### SOLUZIONE 
devo normalizzare i dati su cui calcolo i cluster.<div></div>
normalizzo il valore dei giorni: divido tutti i valori per il valore massimo <div></div>
normalizzo lat e longit rispetto alla finestra di valori "valore_massimo" e "valore_minimo" : <div></div>sottraggo ad ogni valore ,"valore_minimo" quindi divido il risultato per ( "valore_massimo" - "valore_minimo" )

```python
# normalizzo i giorni rispetto il loro valore massimo
df['data_modif']=df['data_modif']/df['data_modif'].max()

# normalizzo la longitudine rispetto la sua finestra
longitudine_Max=df.longitudine.max()
longitudine_min=df.longitudine.min()
df['longitudine_norm']=(df['longitudine']-longitudine_min)/(longitudine_Max-longitudine_min)

# normalizzo la latitudine rispetto la sua finestra
latitudine_Max=df.latitudine.max()
latitudine_min=df.latitudine.min()
df['latitudine_norm']=(df['latitudine']-latitudine_min)/(latitudine_Max-latitudine_min)
df.head()
```

```python
colonne_da_selezionare=['data_modif','longitudine_norm','latitudine_norm']
X=df[colonne_da_selezionare]
```

```python
# K-means con 6 cluster
from sklearn.cluster import KMeans
km = KMeans(n_clusters=6, random_state=1)
km.fit(X)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# creazione canvas (le misure sono in pollici))
plt.figure(figsize=(20, 20))

# salva il DataFrame dei centri dei cluster
df['cluster']=km.labels_
centers = df.groupby('cluster').mean()

# crea un array colors per il grafico
colors = np.array(['red', 'green', 'blue', 'yellow', 'black', 'orange'])

# grafico a dispersione 
# colorato per cluster (0=rosso, 1=verde, 2=blu)
plt.scatter(df.longitudine, df.latitudine, c=colors[list(df.cluster)], s=30)

# centri dei cluster, indicati con "+"
plt.scatter(centers.longitudine, centers.latitudine, linewidths=100, marker='+', s=500,c='black')

# aggiunge le etichette
plt.xlabel('latitudine')
plt.ylabel('longitudine')
```
